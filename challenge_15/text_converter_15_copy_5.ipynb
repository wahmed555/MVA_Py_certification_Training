```python
import csv
# 1. you have to open a file in correct mode before reading from file
fileName='abbre.csv'
accessMode='r'
fss=[]
# open myfile
with open(fileName,accessMode) as myfile:
    allRowList=csv.reader(myfile)

#     now appending each line from allRowList  to  a (list of List) fss
    for currentRow in allRowList:
        fss.append(currentRow)
        print(currentRow)

print(fss)
```

```python
# Now exracting two  string of each string/line (list) from (list of list) using nested loop
nss=[]
for fs in fss:
    for f in fs:
        ns=f.split('-')   # spliting single string list into two string list
    nss.append(ns)
# len(f)  
```
